```python
import requests
import urllib
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt

### WEB SCRAPPER - hepsiburada ###

# Build a string the main URL, search term, and location
def get_url(query):
    url = 'https://www.cimri.com/'
    query = '' + urllib.parse.quote_plus(query)
    url += query
    return url

# Request HMTL from URL, create and return a SOUP object with it 
def get_soup(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.76 Safari/537.36'}
    page = requests.get(url, headers=headers)
    return BeautifulSoup(page.content, 'html.parser')
```

```python
regex = re.compile('.*z7ntrt-0 cLlfW.*')
cards = soup.find_all(class_=regex)

product_title, top_offers, best_offer = [], [], []
for card in cards:
    if str(card.find(class_='product-title')) != 'None':
        product_title.append(card.find(class_='product-title').text)
        top_offers.append(card.find(class_='top-offers').text)
        
        if str(card.find(class_='s14oa9nh-0 fFCyge')) != 'None':
               best_offer.append(card.find(class_='s14oa9nh-0 fFCyge').text)
        else:
            
               best_offer.append('None')
```

```python
query = 'lenovo'

for start in np.arange(0,31,1):
    url= get_url(query)+ '?page=' + str(start) + '&sort=rank%2Cdesc'
    soup = get_soup(url)
    
    regex = re.compile('.*z7ntrt-0 cLlfW.*')
    cards = soup.find_all(class_=regex)
    
    if start == 0:
        product_title, top_offers, best_offer = [], [], []
    for card in cards:
        if str(card.find(class_='product-title')) != 'None':
            product_title.append(card.find(class_='product-title').text)
            top_offers.append(card.find(class_='top-offers').text)
        
            if str(card.find(class_='s14oa9nh-0 fFCyge')) != 'None':
                   best_offer.append(card.find(class_='s14oa9nh-0 fFCyge').text)
            else:
            
                  best_offer.append('None')


```

```python
df = pd.DataFrame(product_title, columns=['product_title'])

df['top_offers'], df['best_offer'] = top_offers, best_offer

df
```
